```python
import time
import os
import re
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from tqdm import tqdm
import pandas as pd
import requests

item = input()
# Driver creation
chrome_options = Options() # Initializing am options object 

chrome_options.add_argument("--no-sandbox")
# Silent download of drivers
os.environ['WDM_LOG'] = 'False'

# Create service
webdriver_service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service = webdriver_service, options = chrome_options)
price = []
name = []
rating = []
brand = []
desc = []
lin = "https://www.amazon.in/s?k=" + item
driver.get(lin)
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')
b = soup.find('span',class_="s-pagination-item s-pagination-disabled")
link = "https://www.amazon.in/s?k=" + item 
driver.get(link)
page_source = driver.page_source
        # Parse 1-page
soup = BeautifulSoup(page_source, 'lxml')
cat = soup.find('div',id="departments")
cate = cat.find_all('span',class_="a-size-base a-color-base")
cate_lin = cat.find_all('a',class_="a-link-normal s-navigation-item")
category = []
categorylinks = []
for i in cate:
        category.append(i.text)
for j in cate_lin:
        categorylinks.append("https://www.amazon.in"+j.get('href'))
catdict = dict(zip(category,categorylinks))
print(category)
num = int(input("Enter the category number you want to search for : "))
driver.get(catdict[category[num-1]])
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')
rev = ['4 stars and above','3 stars and above','2 stars and above', '1 star and above']
revi = soup.find('div',id="reviewsRefinements")
kk = "https://amazon.in"
revie = revi.find_all('a',class_="a-link-normal s-navigation-item")
revlinks = []
for i in revie:
        f = kk + i.get('href')
        revlinks.append(f)
nums = int(input("Customer Reviews should be at least how many stars on a scale of 1 to 4: "))
driver.get(revlinks[4-nums])
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')
pr = soup.find('div',id="priceRefinements")
pric = pr.find_all('span',class_="a-size-base a-color-base")
priclink = pr.find_all('a',class_="a-link-normal s-navigation-item")
prices = []
pricelink = []
for i in pric:
        prices.append(i.text)
for i in priclink:
        pricelink.append(kk+i.get('href'))
print(prices)
selectpricerange = int(input("Which price range do you wish to select? : "))
driver.get(pricelink[selectpricerange])
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'lxml')
bran = soup.find('div',id="brandsRefinements")
brand = bran.find_all('span',class_="a-size-base a-color-base")
branlinks = bran.find_all('a',class_="a-link-normal s-navigation-item")
brandname = []
brandlink = []
for i in brand:
        brandname.append(i.text)
for i in branlinks:
        brandlink.append(kk+i.get('href'))

print(brandname)
g = int(input("How many brands you wish to select? "))
l = []
for i in range(g):
        j = int(input(""))
        l.append(j)
f = 1
pricess = []
rating = []
name = []
if g==0:
        l = [i for i in range(len(brandname))]
for i in l:
        driver.get(brandlink[i])
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'lxml')
        n = soup.find_all('span',class_="a-size-base-plus a-color-base a-text-normal")
        r = soup.find_all('span',class_="a-icon-alt")
        pri = soup.find_all('span',class_="a-price-whole")
        for i in pri:
                pricess.append(i.text)
        for i in n:
                if len(pricess)>len(name):
                        name.append(i.text)
        for i in r:
                p = i.text
                g = p.split(" ")
                if len(pricess)>len(rating):
                        rating.append(g[0])

        m = soup.find('a',class_="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator")
        while (m!=None):
                driver.get(kk+m.get('href'))
                page_source = driver.page_source
                soup = BeautifulSoup(page_source, 'lxml')
                n = soup.find_all('span',class_="a-size-base-plus a-color-base a-text-normal")
                na = soup.find_all('h2',class_="a-size-mini a-spacing-none a-color-base s-line-clamp-2")
                r = soup.find_all('span',class_="a-icon-alt")
                pri = soup.find_all('span',class_="a-price-whole")
                for i in pri:
                        pricess.append(i.text)
                for i in n:
                        if len(pricess)>len(name):
                              name.append(i.text)
                for o in na:
                        name.append(o.text)
                        print(o)
                for i in r:
                        p = i.text
                        g = p.split(" ")
                        if len(pricess)>len(rating):
                              rating.append(g[0])
                m = soup.find('a',class_="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator")

        
```
